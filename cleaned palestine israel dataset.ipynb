```python
df.fillna(0,inplace=True)
df
```

```python
df['Palestinians Injuries']=df['Palestinians Injuries'].replace("(incl. Jun)",np.nan)
df['Israelis Injuries']=df['Israelis Injuries'].replace("(incl. Jun)",np.nan)
df['Palestinians Injuries']=df['Palestinians Injuries'].replace("(incl. Aug)",np.nan)
df['Israelis Injuries']=df['Israelis Injuries'].replace("(incl. Aug)",np.nan)
```

```python
df = df.iloc[0:249]
df
```

```python
df=df.astype({'Year':int,'Palestinians Injuries':int,'Israelis Injuries':int,'Palestinians Killed':int,'Israelis Killed':int})
df
```

```python
df["Palestinians Casualties"] = (df["Palestinians Killed"]+ df["Palestinians Injuries"])
df["Israelis Casualties"] = (df["Israelis Killed"]+ df["Israelis Injuries"])
df
```

```python
# to save it to a dataframe
df.to_csv("palestine_israel.csv", index = False)
```
